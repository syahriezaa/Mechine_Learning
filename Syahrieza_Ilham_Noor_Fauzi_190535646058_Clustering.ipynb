```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, davies_bouldin_score
import numpy as np
```

```python
X, y_true = make_blobs(n_samples=600, 
                       centers=5, 
                       n_features=3, 
                       cluster_std=0.5, 
                       random_state=58)
plt.scatter(X[:, 0], X[:, 1], s=2);

score=[0]*10
```

dikarenakan 3 angka paling belakang nim saya diawali dengan angka 0 dan tidak memungkinkan untuk di jadikan parameter maka saya hanya memasukan 2 angka terakhir

```python
kmeans1 = KMeans(n_clusters=2)
kmeans1.fit(X)
y_kmeans1 = kmeans1.predict(X)
```

```python
centers = kmeans1.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans1, s=5, cmap='jet')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=250, alpha=0.5);
score[1]=davies_bouldin_score(X,kmeans1.labels_)
print(score[1])
```

```python
kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(X)
y_kmeans2 = kmeans2.predict(X)
```

```python
centers = kmeans2.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans2, s=5, cmap='jet')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=250, alpha=0.5);
kmeans2.n_iter_
score[2]=davies_bouldin_score(X,kmeans2.labels_)
print(score[2])
```
